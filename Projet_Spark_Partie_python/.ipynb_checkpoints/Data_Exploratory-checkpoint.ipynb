### Les attributs

* **sex** : masculin (0) ou féminin (1) ; (Nominal)
* **age** : âge du patient ; (Continu - Bien que les âges enregistrés soient tronqués à des nombres entiers, le concept d'âge est continu)
* **currentSmoker** : si le patient est fumeur actuel ou non (Nominal)
* **cigsPerDay** : nombre moyen de cigarettes que la personne fumait en une journée (peut être considéré comme continu car on peut avoir n'importe quel nombre de cigarettes, même une demi-cigarette)
* **BPMeds** : si le patient prenait ou non des médicaments pour la pression artérielle (Nominal)
* **prevalentStroke** : si le patient avait déjà eu un AVC (Nominal)
* **prevalentHyp** : si le patient était hypertendu (Nominal)
* **diabetes** : si le patient était diabétique ou non (Nominal)
* **totChol** : taux de cholestérol total (Continu)
* **sysBP** : pression artérielle systolique (Continu)
* **diaBP** : pression artérielle diastolique (Continu)
* **BMI** : indice de masse corporelle (Continu)
* **heartRate** : rythme cardiaque (Continu - En recherche médicale, des variables telles que le rythme cardiaque, bien qu'en réalité discrètes, sont considérées comme continues en raison d'un grand nombre de valeurs possibles)
* **glucose** : taux de glucose (Continu)
* **10 year risk of coronary heart disease CHD** : risque de maladie coronarienne sur 10 ans (binaire : "1" signifie "Oui", "0" signifie "Non") - Variable cible

### Categorical Features

```python
categorical_features = ['sex', 'education', 'is_smoking', 'BPMeds', 'prevalentStroke', 'prevalentHyp', 'diabetes']
```

```python
for feature in categorical_features:
    print(feature,':')
    print(data[feature].value_counts())
    print("-----------------")
```

```python
num_plots = len(categorical_features)
total_cols = 2
total_rows = num_plots//total_cols + 1
fig, axs = plt.subplots(nrows=total_rows, ncols=total_cols,
                        figsize=(10, 12), facecolor='w', constrained_layout=True)
for i, var in enumerate(categorical_features):
    row = i//total_cols
    pos = i % total_cols
    plot = sns.countplot(x=var, data=data, ax=axs[row][pos])
```

Parmi les caractéristiques catégoriques :

* `BPmeds`, `prevalentStroke` et `diabetes` sont fortement déséquilibrés.
* Il y a quatre niveaux `d'éducation`, tandis que les autres caractéristiques catégoriques sont toutes binaires.
* Le nombre de fumeurs et de non-fumeurs dans `is_smoking` est presque égal.

### Numerical Features

```python
numeric_features = ['cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']
for feature in numeric_features:
    plt.figure(figsize=(18, 10), facecolor='w')
    sns.distplot(data[feature])
    plt.title('{} Distribution'.format(feature), fontsize=20)
    plt.show()
```

Parmi les caractéristiques numériques :

* `totChol`, `sysBP`, `diaBP` et `BMI` ont une distribution uniforme, tandis que les autres caractéristiques sont inégalement réparties.
* `cigsPerDay` a une distribution très inégale, avec la majorité des données concentrées sur 0.
* `cigsPerDay` et `sysBP` montrent respectivement une forte et une légère asymétrie à droite.

```python
#Distribution of outcome variable, Heart Disease
plt.figure(figsize=(12, 10), facecolor='w')
plt.subplots_adjust(right=1.5)
plt.subplot(121)
sns.countplot(x="TenYearCHD", data=data)
plt.title("Count distribution of TenYearCHD", size=20)
plt.subplot(122)
labels=[0,1]
plt.pie(data["TenYearCHD"].value_counts(),autopct="%1.1f%%",labels=labels,colors=["lime","red"])
plt.show()
```

La distribution est fortement déséquilibrée. En effet, le nombre de cas négatifs est nettement supérieur au nombre de cas positifs. Cela entraînera un problème de déséquilibre de classes lors de l'ajustement de nos modèles. Par conséquent, ce problème doit être abordé et pris en charge.

### Bivariate Analysis

#### Quel genre présente un plus grand risque de maladie coronarienne (CHD) ?

```python
graph_2 = data.groupby("sex", as_index=False).TenYearCHD.sum()
```

```python
#Ploting the above values

plt.figure(figsize=(10,6), facecolor='w')
sns.barplot(x=graph_2["sex"], y=graph_2["TenYearCHD"])
plt.title("Graph showing which gender has more risk of coronary heart disease CHD", size=20)
plt.xlabel("Gender\n0 is female and 1 is male",size=20)
plt.ylabel("TenYearCHD cases", size=20)
plt.xticks(size=12)
plt.yticks(size=12)
```

Selon ce jeu de données, les hommes ont montré un risque légèrement plus élevé de maladie coronarienne TenYearCHD.

```python
#Distribution of current smokers with respect to age
plt.figure(figsize=(30,15), facecolor='w')
sns.countplot(x="age",data=data,hue="is_smoking")
plt.title("Graphique montrant quel groupe d'âge compte le plus de fumeurs.", size=30)
plt.xlabel("age", size=20)
plt.ylabel("age Count", size=20)
plt.xticks(size=12)
plt.yticks(size=12)
```

```python
graph_2 = data.groupby("sex", as_index=False).is_smoking.sum()
```

```python
#Ploting the above values

plt.figure(figsize=(12,8), facecolor='w')
sns.barplot(x=graph_2["sex"], y=graph_2["is_smoking"])
plt.title("Graph showing which gender has more smokers", size=20)
plt.xlabel("Gender\n0 is female and 1 is male",size=20)
plt.ylabel("smoking cases", size=20)
plt.xticks(size=12)
plt.yticks(size=12)
```

On constate que les hommes fument plus que les femmes.
